## Get the top 10 directors with most movies directed and use a boxplot for their gross earnings

## Which of these genres are the most profitable? Plot their sales using different histograms, superimposed in the same axis.

- Romance
- Comedy
- Action
- Fantasy

```python
sales_per_genre = data.groupby('genres').sum().reset_index()[['genres', 'sales']]
sales_per_genre

selected_genres = ['Romance', 'Comedy', 'Action', 'Fantasy']
selected_genres_sales = sales_per_genre[sales_per_genre['genres'].isin(selected_genres)].sort_values(by='sales', ascending=False)


ax = sns.histplot(data[data['genres'] == 'Romance']['sales'], color="red", label="Romance", kde=True, stat="density", linewidth=0)
sns.histplot(data[data['genres'] == 'Comedy']['sales'], color="blue", label="Comedy", kde=True, stat="density", linewidth=0, ax=ax)
sns.histplot(data[data['genres']== 'Action']['sales'], color="teal", label="Action", kde=True, stat="density", linewidth=0, ax=ax)
sns.histplot(data[data['genres'] == 'Fantasy']['sales'], color="black", label="Fantasy", kde=True, stat="density", linewidth=0, ax=ax)
ax.legend()


selected_genres_sales

# Overall, Comedy is the most profitable genre but if the basis is per movie,
# then there is a higher possibility that a Fantasy movie can profit more 
```

## Min-Max Normalization 

Normalization is a technique often applied as part of data preparation for machine learning. The goal of normalization is to change the values of numeric columns in the dataset to a common scale, without distorting differences in the ranges of values. For machine learning, every dataset does not require normalization. It is required only when features have different ranges.

The min-max approach (often called normalization) rescales the feature to a hard and fast range of [0,1] by subtracting the minimum value of the feature then dividing by the range. We can apply the min-max scaling in Pandas using the .min() and .max() methods.

$$ 
x_{scaled} = \frac{x-x_{min}}{x_{max}-x_{min}}
$$

### Normalize each numeric column (those that have types integer or float) of the copied dataframe (df)

```python
df.dtypes
```

```python
column_type = "" # Variable to store column type

for (column_name, column_data) in df.iteritems():
    column_type = df[column_name].dtype
    if (column_type == 'float64' or column_type == 'int64' and column_name != 'Unnamed: 0'):
        max = df[column_name].max()
        min = df[column_name].min()
        df[column_name] = (df[column_name] - min) / (max-min)

df
```
